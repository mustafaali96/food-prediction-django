```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df = df[['name', 'category', 'foodCategory', 'Ingredient__ingredient']]
df.rename(columns = {'Ingredient__ingredient':'ingredient'}, inplace = True)
```

```python
df['value'] = 1
```

```python
df
```

```python
df = np.round(pd.pivot_table(df, values='value',
                                index=['name', 'category', 'foodCategory'], 
                                columns= 'ingredient', 
                                aggfunc=np.mean,
                                fill_value=0),2)
```

```python
df
```

```python
df.reset_index(inplace=True)
```

```python
df
```

```python
print(df.columns.name)
```

```python
df.columns.name = None
```

```python
df
```

```python
X = df.loc[:, df.columns!='name']
y = df["name"]
```

```python
from sklearn.tree import DecisionTreeClassifier 
clf = DecisionTreeClassifier()
# Train Decision Tree Classifer
clf = clf.fit(X,y)
```

```python
X.iloc[1].values
```

```python
clf.predict([X.iloc[3].values])
```

```python
clf.feature_names_in_
```

```python
import pickle

filename = 'finalized_model.sav'
pickle.dump(clf, open(filename, 'wb'))
```
